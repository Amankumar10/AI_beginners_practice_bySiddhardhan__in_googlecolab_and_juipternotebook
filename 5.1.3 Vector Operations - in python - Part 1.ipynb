```python
import numpy as np 
import matplotlib.pylab as plt
import seaborn as sns
```

```python
sns.set()
```

```python
plt.quiver(0,0,4,5)
plt.show()
```

```python
plt.quiver(0,0,4,5, scale_units = 'xy',angles='xy',scale=1)  
plt.xlim(-8,8)
plt.ylim(-8,8)
plt.show()
```

```python
plt.quiver(0,0,4,5, scale_units = 'xy',angles='xy',scale=1,color='b')  
plt.quiver(0,0,-3,-5, scale_units = 'xy',angles='xy',scale=1,color='g')  

plt.xlim(-8,8)
plt.ylim(-8,8)
plt.show()
```
