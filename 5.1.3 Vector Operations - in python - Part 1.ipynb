```python
import numpy as np 
import matplotlib.pylab as plt
import seaborn as sns
```

```python
sns.set()
```

```python
plt.quiver(0,0,4,5)
plt.show()
```

```python
plt.quiver(0,0,4,5, scale_units = 'xy',angles='xy',scale=1)  
plt.xlim(-8,8)
plt.ylim(-8,8)
plt.show()
```

```python
plt.quiver(0,0,4,5, scale_units = 'xy',angles='xy',scale=1,color='b')  
plt.quiver(0,0,-3,-5, scale_units = 'xy',angles='xy',scale=1,color='g')  

plt.xlim(-8,8)
plt.ylim(-8,8)
plt.show()
```

Addition of two vectors

```python
vectors_1 = np.asarray([0,0,2,3])

vectors_2 = np.asarray([0,0,3,-2])
sum = vectors_1  + vectors_2
print(f"Sum of two vectors: {sum}")
```

```python
plt.quiver(0,0,2,3, scale_units = 'xy',angles='xy',scale=1,color='b')  
plt.quiver(0,0,3,-2, scale_units = 'xy',angles='xy',scale=1,color='r')
plt.quiver(0,0,5,1, scale_units = 'xy',angles='xy',scale=1,color='g')  
plt.xlim(-8,8)
plt.ylim(-8,8)
plt.show()
```

Subtraction of two vectors

```python
vectors_1 = np.asarray([0,0,2,3])

vectors_2 = np.asarray([0,0,3,-2])
diff = vectors_1  - vectors_2
print(diff)
```

```python
plt.quiver(0,0,2,3, scale_units = 'xy',angles='xy',scale=1,color='b')  
plt.quiver(0,0,3,-2, scale_units = 'xy',angles='xy',scale=1,color='r')
plt.quiver(0,0,-1,5, scale_units = 'xy',angles='xy',scale=1,color='g')  
plt.xlim(-8,8)
plt.ylim(-8,8)
plt.show()
```
